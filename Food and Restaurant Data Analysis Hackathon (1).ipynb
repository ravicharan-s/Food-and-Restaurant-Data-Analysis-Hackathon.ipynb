# Data Cleaning & Preprocessing

### Step 1: Rating Column (rate)

- Replace '-' values with NaN


- Remove /5 and retain only the numeric value


- Convert to float using pd.to_numeric()


- Fill missing values using the median rating

```python
df["rate"]
```

```python
set(df["rate"])
```

```python
#Replace '-' values with NaN
df["rate"]=df["rate"].replace("-",np.nan)
```

```python
# removing "NEW"
df["rate"]=df["rate"].str.replace("NEW","")
```

```python
#Remove /5 and retain only the numeric value
df["rate"]=df["rate"].str.replace("/5","")
```

```python
#Convert to float using pd.to_numeric()
df["rate"]=pd.to_numeric(df["rate"])
df["rate"].dtype
```

```python
# Fill missing values using the median rating
df["rate"]=df["rate"].fillna(df["rate"].median())
df["rate"].isna().sum()
```

### Step 2: Cost Column (approx_costfor_two_people)

- Remove commas from numeric strings ('1,000' → '1000')


- Convert to numeric


- Fill missing values using the median cost

```python
df["approx_costfor_two_people"]
```

```python
#Remove commas from numeric strings ('1,000' → '1000')
df["approx_costfor_two_people"]=df["approx_costfor_two_people"].str.replace(",","")
```

```python
#Convert to numeric
df["approx_costfor_two_people"]=pd.to_numeric(df["approx_costfor_two_people"])
df["approx_costfor_two_people"].dtype
```

```python
#Fill missing values using the median cost
df["approx_costfor_two_people"]=df["approx_costfor_two_people"].fillna(df["approx_costfor_two_people"].median())
df["approx_costfor_two_people"].isna().sum()
```

### Step 3: Categorical Columns

- dish_liked → Replace NaN with "Not Available"


- cuisines → Replace NaN with "Other"


- rest_type → Replace NaN with "Unknown"

```python
set(df["dish_liked"])
```

```python
#dish_liked → Replace NaN with "Not Available"
df["dish_liked"]=df["dish_liked"].replace(np.nan,"Not Available")
df["dish_liked"].isna().sum()
```

```python
set(df["cuisines"])
```

```python
#cuisines → Replace NaN with "Other"
df["cuisines"]=df["cuisines"].replace(np.nan,"Other")
df["cuisines"].isna().sum()
```

```python
set(df["rest_type"])
```

```python
#rest_type → Replace NaN with "Unknown"
df["rest_type"]=df["rest_type"].replace(np.nan,"Other")
df["rest_type"].isna().sum()
```

### Step 5: Binary Encoding
- Convert the following binary fields:
- Column             Mapping

 online_order    Yes → 1, No → 0
 
 book_table      Yes → 1, No → 0



```python
set(df["online_order"])
```

```python
#online_order Yes → 1, No → 0
df["online_order"]=df["online_order"].replace("Yes",1)
df["online_order"]=df["online_order"].replace(["NO","No"],0)
set(df["online_order"])
```

```python
set(df["book_table"])
```

```python
#book_table Yes → 1, No → 0
df["book_table"]=df["book_table"].replace("Yes",1)
df["book_table"]=df["book_table"].replace("No",0)
set(df["online_order"])
```

# removing nan values

```python
df2=df2.dropna(subset=["Latitude","Longitude"])
```

```python
import folium
from folium.plugins import HeatMap
```

```python
bangalore_map = folium.Map(location=[12.9716, 77.5946], zoom_start=12)

restaurant_locations = df2[['Latitude', 'Longitude']].values.tolist()

HeatMap(restaurant_locations).add_to(bangalore_map)

bangalore_map.save("restaurant_density_map.html")
```

```python
italian_df = df2[df2['cuisines'].str.contains("Italian", case=False, na=False)]

italian_map = folium.Map(location=[12.9716, 77.5946], zoomstart=12)

for  _, row in italian_df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['rate'],
        icon=folium.Icon(color="red")
    ).add_to(italian_map)

italian_map.save("italian_restaurants_map.html")
```
